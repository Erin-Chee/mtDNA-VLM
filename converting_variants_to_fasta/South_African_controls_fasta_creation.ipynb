```python

Mitokor_sequences_AD=[] # create empty list to convert to fasta??
patient_id=[]
for patient in range(0,len(df)):
    Variants=str(df.loc[patient,:].values).replace("[","").replace("]","").replace("'","").replace(" ","").replace("N","").split(";")
    Variants=[i for i in Variants if not i[0:3]=="AJH" ] # remove this object unsure what it means
    patient_id.append(Variants[0]) # remove patient/sequence id not needed in loop but save for later
    Variants=Variants[1:] # remove patient/sequence id not needed in loop but save for later
    with open("Reference mtDNA.txt", "r") as file:
        single_line = file.read().replace('\n', ' ') # read the sequence as a single line
    Template_sequence=list(single_line.replace(" ",""))
    
    for Var in Variants:
        try: # need to use try and except block due to some lists containing empty elements . Not an issue for applying variants
            Variant_base=Var[-1] 
            #print(Var[-1])
            Current_variant=Var[1:]
            Current_variant=Current_variant.replace(f"{Current_variant[-1]}","")
            #print(Current_variant)
            Current_variant_position=int(Current_variant)-1 # minus 1 so that position matches python indexing
        
            Template_sequence[Current_variant_position]=Variant_base
        except:
            continue
    Mitokor_sequences_AD.append(''.join(Template_sequence).replace("N","")) # some sequences contain "N" unknown character that should be removed
```

```python
# Write final results to a fasta file
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO

records=[]
for seq, id_name in zip(Mitokor_sequences_AD,patient_id):
    record=SeqRecord(Seq(seq),id=id_name,description="") #create a sequence record for each using bioPython
    records.append(record)

SeqIO.write(records,'Mitokor_AD_mtDNA.fasta',format="fasta")
```
