```r
# Normalise tRNA and protein encoding data and add their scores to obtain a combined mtVLM score

robust_z_score <- function(x) {
  med <- median(x, na.rm = TRUE)
  iqr <- IQR(x, na.rm = TRUE)
  (x - med) / iqr
}


phenotype<-read.csv("CFS_Samples_UK_with_mtVLM.csv")
APOGEE_UK_Fatigue_PE<-read.csv("results (3).csv")
Patient_split_PE <- lapply(APOGEE_UK_Fatigue_PE$Patient, function(x) strsplit(x, "")[[1]])
APOGEE_UK_Fatigue_tRNA<-read.csv("results (5)_tRNA.csv")
Patient_split_tRNA <- lapply(APOGEE_UK_Fatigue_tRNA$Patient, function(x) strsplit(x, "")[[1]])


# Normalise tRNA scores and PE scores
APOGEE_UK_Fatigue_PE$Variant.Load.Score=robust_z_score(APOGEE_UK_Fatigue_PE$Variant.Load.Score)
APOGEE_UK_Fatigue_tRNA$Variant.Load.Score=robust_z_score(APOGEE_UK_Fatigue_tRNA$Variant.Load.Score)

# Calculate the combined PE and tRNA scores
APOGEE_UK_Fatigue_PE$Total_mtVLM=APOGEE_UK_Fatigue_PE$Variant.Load.Score+APOGEE_UK_Fatigue_tRNA$Variant.Load.Score


#
Patient=c()
for(i in c(1:length(Patient_split_PE))){
collapse_ID=paste(Patient_split_PE[[i]][30:32],collapse = "")
ID=paste("CFS",collapse_ID,sep = "")
Patient=c(Patient,ID)
}
APOGEE_UK_Fatigue_PE$Patient=Patient

# Merge fatigue data with APOGEE variant load scores
df=merge(APOGEE_UK_Fatigue_PE,phenotype,by="Patient")

Column_names=names(phenotype)[7:25]

df_corr_TOTAL_NORM=data.frame(Correlation=0,Phenotype="NA",Sample_size=0)
for (i in Column_names){
Pheno_data=na.omit(df[,c("Patient","Total_mtVLM",i)])
correlation=cor(Pheno_data$Total_mtVLM,Pheno_data[i],method="pearson")
#print(paste(i,"and APOGEE correlation",correlation,"n=",length(row.names(Pheno_data[i]))))
df_corr_TOTAL_NORM=rbind(df_corr_TOTAL_NORM,data.frame(Correlation=correlation[1],Phenotype=i,Sample_size=length(row.names(Pheno_data[i]))))
}
df_corr_TOTAL_NORM['APOGEE']=replicate(length(row.names(df_corr_TOTAL_NORM)),"mtVLM (APOGEE tRNA + PE normalised)")
df_corr_TOTAL_NORM[1:5,]
```
