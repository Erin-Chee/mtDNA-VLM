```python
install.packages("cowplot")
library("cowplot")
library(ggplot2)
```

```python
unzip("AD_results.zip", exdir = "./") # unzip all results
unzip("Japanese_results.zip", exdir = "./") # unzip all results
```

```python
# MEAN SCORES AD
AD_MUTPRED=read.csv("/content/AD_results/Protein encoding results/MutPred_scores/results_Mutpred_scores_AD.csv")
AD_MUTPRED_50=read.csv("/content/AD_results/Protein encoding results/MutPred_0.5_threshold_scores/results_Mutpred_scores_AD_50_threshold.csv")
AD_MUTPRED_70=read.csv("/content/AD_results/Protein encoding results/MutPred_0.7_threshold_scores/mut_pred_AD_threshold_70.csv")

AD_MUTPRED_MEAN=mean(AD_MUTPRED$Variant.Load.Score)
AD_MUTPRED_50_MEAN=mean(AD_MUTPRED_50$Variant.Load.Score)
AD_MUTPRED_70_MEAN=mean(AD_MUTPRED_70$Variant.Load.Score)#


# MEAN SCORES Japanese
JAP_MUTPRED=read.csv("/content/Japanese_results/MutPred_scores/results_Mutpred_scores_Japenese.csv")
JAP_MUTPRED_50=read.csv("/content/Japanese_results/MutPred_0.5_threshold_scores/results_Mutpred_scores_Japenese_50_threshold.csv")
JAP_MUTPRED_70=read.csv("/content/Japanese_results/MutPred_0.7_threshold_scores/results_Mutpred_scores_Japenese_70_threshold.csv")

JAP_MUTPRED_MEAN=mean(JAP_MUTPRED$Variant.Load.Score)
JAP_MUTPRED_50_MEAN=mean(JAP_MUTPRED_50$Variant.Load.Score)
JAP_MUTPRED_70_MEAN=mean(JAP_MUTPRED_70$Variant.Load.Score)
```

```python
# MAKE dataframe for heatmap

labels=c("Mean variant load (MutPred)","Mean variant load (MutPred >0.5)","Mean variant load (MutPred >0.7)","Mean variant load (MutPred)","Mean variant load (MutPred >0.5)","Mean variant load (MutPred >0.7)")
cohort=c("Alzheimer's","Alzheimer's","Alzheimer's","Centenarians","Centenarians","Centenarians")
means=c(AD_MUTPRED_MEAN,AD_MUTPRED_50_MEAN,AD_MUTPRED_70_MEAN,JAP_MUTPRED_MEAN,JAP_MUTPRED_50_MEAN,JAP_MUTPRED_70_MEAN)

df_heatmap=data.frame(MutPred=labels, Mean_variant_load=means,cohort=cohort)
df_heatmap
```

```python

# subset_APOGEE_scores=APOGEE_scores[0:10,]


plot_MutPred=ggplot(df_heatmap, aes(x = cohort, y = MutPred, fill = Mean_variant_load)) +
  geom_tile(color = "black") +
  scale_fill_gradient(low = "white", high = "darkblue") +
  coord_fixed()+xlab("")+theme_classic()
plot_MutPred
```

```python
# MEAN SCORES AD
AD_APOGEE=read.csv("/content/AD_results/Protein encoding results/APOGEE_scores/results.csv")
AD_APOGEE_20=read.csv("/content/AD_results/Protein encoding results/APOGEE_0.265_threshold_scores/results (1).csv")
AD_APOGEE_70=read.csv("/content/AD_results/Protein encoding results/APOGEE_0.716_threshold_scores/results (1).csv")

AD_APOGEE_MEAN=mean(AD_APOGEE$Variant.Load.Score)
AD_APOGEE_20_MEAN=mean(AD_APOGEE_20$Variant.Load.Score)
AD_APOGEE_70_MEAN=mean(AD_APOGEE_70$Variant.Load.Score)#


# MEAN SCORES Japanese
JAP_APOGEE=read.csv("/content/Japanese_results/APOGEE_scores/results.csv")
JAP_APOGEE_20=read.csv("/content/Japanese_results/APOGEE_0.265_threshold_scores/results (1).csv")
JAP_APOGEE_70=read.csv("/content/Japanese_results/APOGEE_0.716_threshold_scores/results (1).csv")

JAP_APOGEE_MEAN=mean(JAP_APOGEE$Variant.Load.Score)
JAP_APOGEE_20_MEAN=mean(JAP_APOGEE_20$Variant.Load.Score)
JAP_APOGEE_70_MEAN=mean(JAP_APOGEE_70$Variant.Load.Score)

# MAKE dataframe for heatmap

labels=c("Mean variant load (APOGEE)","Mean variant load (APOGEE >0.265)","Mean variant load (APOGEE >0.716)","Mean variant load (APOGEE)","Mean variant load (APOGEE >0.265)","Mean variant load (APOGEE >0.716)")
cohort=c("Alzheimer's","Alzheimer's","Alzheimer's","Centenarians","Centenarians","Centenarians")
means=c(AD_APOGEE_MEAN,AD_APOGEE_20_MEAN,AD_APOGEE_70_MEAN,JAP_APOGEE_MEAN,JAP_APOGEE_20_MEAN,JAP_APOGEE_70_MEAN)

df_heatmap=data.frame(APOGEE=labels, Mean_variant_load=means,cohort=cohort)
df_heatmap
```

```python

plot_APOGEE=ggplot(df_heatmap, aes(x = factor(cohort), y = APOGEE, fill = Mean_variant_load)) +
  geom_tile(color = "black") +
  scale_fill_gradient(low = "white", high = "darkred") +
  coord_fixed()+xlab("")+theme_classic()
plot_APOGEE
```
